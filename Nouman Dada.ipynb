```python
import math

# Minimax function
def minimax(depth, nodeIndex, maximizingPlayer, values, targetDepth):

    # Base case: if leaf node reached
    if depth == targetDepth:
        return values[nodeIndex]

    # Maximizing player's turn
    if maximizingPlayer:
        left = minimax(depth + 1, nodeIndex * 2, False, values, targetDepth)
        right = minimax(depth + 1, nodeIndex * 2 + 1, False, values, targetDepth)

        best = max(left, right)
        print(f"MAX chooses max({left}, {right}) = {best}")
        return best

    # Minimizing player's turn
    else:
        left = minimax(depth + 1, nodeIndex * 2, True, values, targetDepth)
        right = minimax(depth + 1, nodeIndex * 2 + 1, True, values, targetDepth)

        best = min(left, right)
        print(f"MIN chooses min({left}, {right}) = {best}")
        return best


# Leaf nodes from your tree
values = [-4, -5, 2, 3, 4, 5, 6, 7]

# Height of tree = 3
treeDepth = 3

# Run minimax
result = minimax(0, 0, True, values, treeDepth)

print("\nOptimal value is:", result)


import math

def alpha_beta(depth, nodeIndex, maximizingPlayer, values, alpha, beta, targetDepth):

    # Base case → leaf node
    if depth == targetDepth:
        return values[nodeIndex]

    # MAX player
    if maximizingPlayer:
        best = -math.inf

        for i in range(2):   # left child, right child
            val = alpha_beta(depth + 1,
                             nodeIndex * 2 + i,
                             False,
                             values,
                             alpha,
                             beta,
                             targetDepth)

            best = max(best, val)
            alpha = max(alpha, best)

            print(f"MAX: best={best}, alpha={alpha}, beta={beta}")

            # Pruning
            if beta <= alpha:
                print("PRUNED at MAX")
                break

        return best

    # MIN player
    else:
        best = math.inf

        for i in range(2):
            val = alpha_beta(depth + 1,
                             nodeIndex * 2 + i,
                             True,
                             values,
                             alpha,
                             beta,
                             targetDepth)

            best = min(best, val)
            beta = min(beta, best)

            print(f"MIN: best={best}, alpha={alpha}, beta={beta}")

            # Pruning
            if beta <= alpha:
                print("PRUNED at MIN")
                break

        return best


# Leaf nodes
values = [4, 3, 2, 5, 0, 1, 6, 5]

# Tree depth
treeDepth = 3

# Run Alpha-Beta
result = alpha_beta(
    0,              # depth
    0,              # root index
    True,           # MAX starts
    values,
    -math.inf,      # alpha
    math.inf,       # beta
    treeDepth
)

print("\nOptimal value =", result)
```
