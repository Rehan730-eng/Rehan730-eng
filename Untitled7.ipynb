```python

import random

def objective_function(x):

    return -(x**2) + 4

def simple_hill_climbing(iterations, step_size):

    current_x = random.uniform(-10, 10)
    current_val = objective_function(current_x)

    for i in range(iterations):

        neighbor_x = current_x + random.uniform(-step_size, step_size)
        neighbor_val = objective_function(neighbor_x)

        if neighbor_val > current_val:
            current_x = neighbor_x
            current_val = neighbor_val

    return current_x, current_val


best_x, best_val = simple_hill_climbing(100, 0.1)

print("Best X found:", best_x)
print("Best Value:", best_val)
```
